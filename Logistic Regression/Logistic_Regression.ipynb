```python
# Drop Cabin due to too many missing values
data.drop(columns=['Cabin'], inplace=True)
```

```python
# Encoding the columns 'Sex': male â†’ 0, female â†’ 1
#if it is male it displays '0', female it displays '1'
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
```

```python
data['Sex']
```

```python
#using dummies
data = pd.get_dummies(data, columns=['Embarked'], drop_first=True)
data
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

```python
data.isnull().sum()
```

```python
# Feature selection
x=data.loc[:, ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_Q', 'Embarked_S']]
x
```

```python
y = data['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python

y_proba = model.predict_proba(x_test)[:, 1]
y_proba
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
roc_auc_score(y_test, y_proba)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc_score(y_test, y_proba):.2f})", color='green')
plt.plot([0, 1], [0, 1], 'k--')  
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for given data")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python

# Get the coefficients from the trained model
coefficients = model.coef_[0]

# Create a DataFrame showing feature names and their corresponding coefficients
coef_data = pd.DataFrame({
    'Feature': x_train.columns,
    'Coefficient': coefficients
})
```

```python
print(" Logistic Regression Coefficients:")
print(coef_data)
#From coef_data, find features with largest positive or negative coefficients.

#The most significant features are:

#Sex (very strong negative coefficient for males)

#Pclass_3 (3rd class passengers are less likely to survive)

#Fare (wealthier passengers had better chances)
```
